# **Importing the Libraries and reading the CSV File**

---



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/breast_cancer_survival - breast_cancer_survival_.csv.csv')
df.info() #getting an overview about number of rows, columns, null values and datatypes
```

# **Data Preprocessing**

---
Dropping Duplicate Rows.


```python
df = df.drop_duplicates()
df
```

Dropping rows with null values in 'Date_of_Last_Visit' and 'Patient_Status' column.

```python
df.dropna(subset = ['Date_of_Last_Visit', 'Patient_Status'], inplace = True)
df
```

Checking for unique values.

```python
df.nunique()
```

Dropping 'ER Status' and 'PR Status' column because it only had one value and also dropping the 'Date_of_Last_Visit' column because the data is irrelevent to the analysis.

```python
df.drop(['ER status',	'PR status', 'Date_of_Last_Visit'], axis = 1, inplace = True)
```

```python
df.info()
```

Sorting the rows by their Age.

```python
df.sort_values(by = 'Age', inplace = True)
df
```

Converting 'Tumour_Stage' entries from string to integer.

```python
df['Tumour_Stage'] = pd.Categorical(df['Tumour_Stage'], categories=['I', 'II', 'III'], ordered=True).codes + 1
df
```

Resetting the index and duplicating the DataFrame for graph creation.

```python
df = df.reset_index(drop = True)
df2 = df.copy()
df2
```

# **GRAPHS**

---
Getting the count for Age.

```python
sns.histplot(df2['Age'], kde=True)
plt.title('Distribution of Age of Patients')
```

Getting the count for different columns.

Although rare, men do have a small chance of developing breast cancer.

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

sns.countplot(ax=axes[0, 0], x='Tumour_Stage', data=df2)
axes[0, 0].set_title('Distribution of Tumour Stages')

sns.countplot(ax=axes[0, 1], x='Gender', data=df2)
axes[0, 1].set_title('Distribution of Gender')

sns.countplot(ax=axes[1, 0], x='HER2 status', data=df2)
axes[1, 0].set_title('Distribution of HER2 Status')

sns.countplot(ax=axes[1, 1], x='Surgery_type', data=df2)
axes[1, 1].set_title('Distribution of Surgery Types')
plt.xticks(rotation=15)
```

Pie chart for Patient Status and Count Plot for Histology Types.

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 10))

status_counts = df2['Patient_Status'].value_counts()
axes[0].pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', colors=['#66c2a5', '#fc8d62'])
axes[0].set_title('Distribution of Patient Status')

sns.countplot(x="Histology", data=df2, ax=axes[1])
axes[1].tick_params(axis='x', rotation=15)
axes[1].set_title('Count of Histology Types')
```

Box plot for all Proteins with respect to each other and Age distribution by HER2 Status.

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

sns.boxplot(data=df2[['Protein1', 'Protein2', 'Protein3', 'Protein4']], ax=axes[0])
axes[0].set_title('Distribution of Protein Levels Across Different Proteins')

sns.boxplot(data=df, x='HER2 status', y='Age', ax=axes[1])
axes[1].set_title('Age Distribution by HER2 Status')
```

Violin plot for all Proteins vs Tumour Stage.

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10), sharex=True)

sns.violinplot(ax=axes[0, 0], x='Tumour_Stage', y='Protein1', data=df2)
axes[0, 0].set_title('Protein1 vs Tumour Stage')

sns.violinplot(ax=axes[0, 1], x='Tumour_Stage', y='Protein2', data=df2)
axes[0, 1].set_title('Protein2 vs Tumour Stage')

sns.violinplot(ax=axes[1, 0], x='Tumour_Stage', y='Protein3', data=df2)
axes[1, 0].set_title('Protein3 vs Tumour Stage')

sns.violinplot(ax=axes[1, 1], x='Tumour_Stage', y='Protein4', data=df2)
axes[1, 1].set_title('Protein4 vs Tumour Stage')
```

Box plot for all Proteins vs Age.

```python
df2['Age_Binned'] = pd.cut(df2['Age'], bins=range(20, df2['Age'].max() + 15, 15), right=False)

fig, axes = plt.subplots(2, 2, figsize=(12, 10), sharex=True)

sns.boxplot(ax=axes[0, 0], x='Age_Binned', y='Protein1', data=df2)
axes[0, 0].set_title('Protein1 vs Age')

sns.boxplot(ax=axes[0, 1], x='Age_Binned', y='Protein2', data=df2)
axes[0, 1].set_title('Protein2 vs Age')

sns.boxplot(ax=axes[1, 0], x='Age_Binned', y='Protein3', data=df2)
axes[1, 0].set_title('Protein3 vs Age')

sns.boxplot(ax=axes[1, 1], x='Age_Binned', y='Protein4', data=df2)
axes[1, 1].set_title('Protein4 vs Age')
```

Box plot for all Proteins vs Gender.

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10), sharex=True)

sns.boxplot(ax=axes[0, 0], x='Gender', y='Protein1', data=df2)
axes[0, 0].set_title('Protein1 vs Gender')

sns.boxplot(ax=axes[0, 1], x='Gender', y='Protein2', data=df2)
axes[0, 1].set_title('Protein2 vs Gender')

sns.boxplot(ax=axes[1, 0], x='Gender', y='Protein3', data=df2)
axes[1, 0].set_title('Protein3 vs Gender')

sns.boxplot(ax=axes[1, 1], x='Gender', y='Protein4', data=df2)
axes[1, 1].set_title('Protein4 vs Gender')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10), sharex=True)

sns.boxplot(data=df, x='HER2 status', y='Protein1', ax=axes[0, 0])
axes[0, 0].set_title('Protein1 Levels by HER2 Status')

sns.boxplot(data=df, x='HER2 status', y='Protein2', ax=axes[0, 1])
axes[0, 1].set_title('Protein2 Levels by HER2 Status')

sns.boxplot(data=df, x='HER2 status', y='Protein3', ax=axes[1, 0])
axes[1, 0].set_title('Protein3 Levels by HER2 Status')

sns.boxplot(data=df, x='HER2 status', y='Protein4', ax=axes[1, 1])
axes[1, 1].set_title('Protein4 Levels by HER2 Status')
```

These visualizations explore the relationship between tumour stages, patient statuses, HER2 status, and surgery types, including distributions and stacked bar plots.

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

sns.countplot(data=df2, x='Tumour_Stage', hue='HER2 status', ax=axes[0, 0])
axes[0, 0].set_title('Tumour Stage by HER2 Status')

sns.countplot(data=df2, x='Patient_Status', hue='HER2 status', ax=axes[0, 1])
axes[0, 1].set_title('Patient Status by HER2 Status')

counts = df2.groupby(['Tumour_Stage', 'Patient_Status']).size().unstack(fill_value=0)
counts.plot(kind='bar', stacked=True, ax=axes[1, 0])
axes[1, 0].set_title('Tumour Stage and Patient Status')
axes[1, 0].tick_params(axis='x', rotation=0)

sns.countplot(data=df, x='Surgery_type', hue='Patient_Status', ax=axes[1, 1])
axes[1, 1].set_title('Surgery Type by Patient Status')
axes[1, 1].tick_params(axis='x', rotation=15)
```
